#### Get Winning sets of states for any player 1 for alu 2 for plus

```python
# get all winning combinations 
def get_winning_sets(p):

    winning_sets = set()

    def won(c, n):
      if c[0] == n and c[1] == n and c[2] == n: return 1
      if c[3] == n and c[4] == n and c[5] == n: return 1
      if c[6] == n and c[7] == n and c[8] == n: return 1

      if c[0] == n and c[3] == n and c[6] == n: return 1
      if c[1] == n and c[4] == n and c[7] == n: return 1
      if c[2] == n and c[5] == n and c[8] == n: return 1

      if c[0] == n and c[4] == n and c[8] == n: return 1
      if c[2] == n and c[4] == n and c[6] == n: return 1

      return 0

    pc = [' ', 'x', 'o']
    c = [0] * 9
    for c[0] in range (3):
      for c[1] in range (3):
        for c[2] in range (3):
          for c[3] in range (3):
            for c[4] in range (3):
              for c[5] in range (3):
                for c[6] in range (3):
                  for c[7] in range (3):
                    for c[8] in range (3):
                      #countx = sum([1 for x in c if x == 1])
                      #county = sum([1 for x in c if x == 2])
                      #if abs(countx-county) < 2:
                        if won(c,p) == 1:
                          sum1 = 0
                          for i in range(9):
                            sum1 += c[i]*(3**i)
                          #print sum1
                          winning_sets.add(sum1)
    return winning_sets
```

  #### Convert state array to a state number

```python
def state_num(state_rep):
    state_num =0 
    for i in range(9):
        state_num += int(state_rep[i])*(3**i)
    return state_num

def ternary (n):
    if n == 0:
        return '0'
    nums = []
    while n:
        n, r = divmod(n, 3)
        nums.append(str(r))
    return ''.join(reversed(nums))

def state_representation(n):
    if n == 0:
        return '000000000'
    nums = []
    while n:
        n, r = divmod(n, 3)
        nums.append(r)
    while len(nums)!=9:
        nums.append(0)
    return nums
```

#### Get next move at random from state_rep for player p

```python
def next_random_move(state_rep,p):
    possible_moves = []
    
    for i in range(9):
        new_move = state_rep[:]
        if state_rep[i]==0:
            new_move[i]=p
            possible_moves.append(new_move[:])
    
    if len(possible_moves) < 1:
        return false
    return(possible_moves[random.randint(0,len(possible_moves)-1)])
```

#### AI Agent

```python
class AluPlusAgent(object):
    def __init__(self,player=1):
        self.valuefunction = {}
        self.possible_moves = []
        self.gameplay = []
        self.p = player
        self.winstates = get_winning_sets(self.p)
        for i in range(19684):
            self.valuefunction.update({i: 0.5})
        for i in self.winstates:
            self.valuefunction.update({i: 1})
        
      
    
    def reset(self):
        self.possible_moves = []
        self.gameplay = []
    
    def next_possible_moves(self,state_rep):
        self.possible_moves = []
        for i in range(9):
            new_move = state_rep[:]
            if state_rep[i]==0:
                new_move[i]=self.p
                self.possible_moves.append(state_num(new_move[:]))
    
        if len(self.possible_moves) < 1:
            return []
        return self.possible_moves
    
    def next_move(self,state_rep):
        available_moves = self.next_possible_moves(state_rep)
        #print "Next Alu Moves"
        #for i in available_moves:
        #    print "available", i, state_representation(i) , self.valuefunction[i]
        max_value = max([v for (k, v) in self.valuefunction.iteritems() if k in available_moves ])
        
        #print "------------------------------------------------"
       # print "max value:", max_value
      #  for (k, v) in self.valuefunction.iteritems():
          #  if ((v == max_value) and (k in available_moves)):
       #       #  print k ,v, self.valuefunction[k]
       # print "------------------------------------------------"
            
        nm = max([k for (k, v) in self.valuefunction.iteritems() if (v == max_value) and (k in available_moves)])
        #print nm, ternary(nm)
        #print state_representation(nm)
        if not available_moves :
            return 0
        else:
            return state_representation(nm)

    
    def next_random_move(self,state_rep):
        self.possible_moves = []
        for i in range(9):
            new_move = state_rep[:]
            if state_rep[i]==0:
                new_move[i]=self.p
                self.possible_moves.append(new_move[:])

        if len(self.possible_moves) < 1:
            return false
        return(self.possible_moves[random.randint(0,len(self.possible_moves)-1)])
    
    def update_value_function(self,win_loss):
        # lambda to simulate the learning curve
        
       #print "New_loop:", len([k for (k, v) in self.valuefunction.iteritems() if v == 0.5  ])

       # print "entered update value function"
        for i in range(len(self.gameplay)):
            #print "updating for state",i
            weight = 0.95
            new_value = win_loss*(weight**(len(self.gameplay)-i-1) )
            update_value =  (self.valuefunction[i]*3 + new_value)/4 
            #print "updating for state",self.gameplay[i], "new_value",new_value, "update_value",update_value
            self.valuefunction.update({self.gameplay[i]: update_value}) 
            #print "updating for state",self.gameplay[i], "new_value",self.valuefunction.get(self.gameplay[i])
            
                
            
```

```python
#print "suru"
Alu_Win = 0
Plus_Win = 0
Draw_Game = 0
last_win = 0
last_draw = 0
Alu = AluPlusAgent(2)
Plus = AluPlusAgent(1)

for episode in range(100000):
    if(episode%20 ==0 and episode!=0):
        
        ep_win = Alu_Win-last_win
        ep_draw = Draw_Game-last_draw
        last_win = Alu_Win
        last_draw = Draw_Game
        print "-------------------------"
        print "Episode Count:", episode 
        print "Win Count:", ep_win, " Draw Count:", ep_draw
        print "Alu Win Rate:", float(Alu_Win)/(episode) 
        print "Win/Draw Rate", float(Draw_Game+Alu_Win)/episode
        print len([k for (k, v) in Alu.valuefunction.iteritems() if v == 0.5  ])
        print "-------------------------"
    #print "Episode:",episode
    Winner = ""
    
    
    
    Alu.reset()
    Plus.reset()

    plus_win_states = get_winning_sets(1)
    alu_win_states = get_winning_sets(2)

    state_rep = [0] * 9
    state = 0


    for i in range (9):
        #time.sleep(1)
        #os.system('clear') 
        #clear_output()
        if(i%2 == 0):
            new_move =Alu.next_move(state_rep)
            #print "optimum play", new_move, Alu.valuefunction[state_num(new_move)]
            #print "actual play", new_move
        else:
            i = random.randint(1,6)
            if(i==1):
                new_move = Plus.next_move(state_rep) 
            else:
                new_move = Plus.next_random_move(state_rep) 

        # make the new_move
        state_rep = new_move

        #make AI remember the gameplay to update the valuefunction at the end 
        Alu.gameplay.append(state_num(state_rep))
        Plus.gameplay.append(state_num(state_rep))

        #print "-----------"
        #print(state_rep,state_num(state_rep))
        
        if((state_num(state_rep) in plus_win_states)):
            print "Plus Win"
            show_board(state_rep) 
            Plus.update_value_function(1)
            Alu.update_value_function(0)
            Winner = "Plus"
            Plus_Win+=1
            #print "Episode:",episode," Winner:", Winner
            break
        if((state_num(state_rep) in alu_win_states) ) :
            print "Alu Win"
            show_board(state_rep) 
            Alu.update_value_function(1)
            Plus.update_value_function(0)
            Winner = "Alu"
            Alu_Win+=1
            #print "Episode:",episode," Winner:", Winner
            break
        if i == 8:
            print "Draw"
            Alu.update_value_function(0.8)
            Plus.update_value_function(0.8)
            Winner = "Draw"
            Draw_Game +=1
            #print "Episode:",episode," Winner:", Winner
        show_board(state_rep) 

print "Final Alu Win Rate:", float(Alu_Win)/(Alu_Win+Plus_Win)
        
        
```
